```python

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd

# 16 weights
```

```python

def activation_function(z):
    return np.where( z>=0, 1, 0  )
```

```python


activation_function( -0.7 )
```

```python

activation_function( 1.7 )
```

```python

def fit(X_train, y_train):

    ##################
    ## Defining the knobs

    weights = np.zeros(  16 )
    bias    = 0
    total_updates = 0
    #################

    learning_rate = 0.01

    for _ in range(1000):
        for idx, x_i in enumerate(X_train):

            linear_product = np.dot(x_i, weights) + bias    ## y = w*x + b   ## Inference Equation
            y_pred  = activation_function(    linear_product    )
            loss =  y_pred   -  y_train[idx]

            if loss != 0:
                total_updates += 1
                update = learning_rate * loss
                weights = weights - update * x_i
                bias    = bias - update


    print(f"Total updates: {total_updates}")
    return weights, bias
```

```python

# generate a 100 samples of each class using claude.ai prompt:

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/WebPerceptron/lt_dataset.csv")

X = df.drop(columns=["class"]).values
y = df["class"].values

X.shape
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=99)
```

```python

weights, bias = fit(X_train, y_train)
```

```python

def predict(X_test, weights, bias):
    ## x * w
    linear_product = np.dot(X_test, weights) + bias    ## y = w*x + b
    y_pred  = activation_function(    linear_product    )
    return y_pred
```

```python

def accuracy(y_true, y_pred):
    accuracy = np.sum(   y_true == y_pred   ) / len(  y_true  )
    return accuracy
```

```python

y_pred = predict(X_test, weights, bias)
```

```python

print(  accuracy(y_pred, y_test)   )
```
